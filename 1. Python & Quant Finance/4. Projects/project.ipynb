```python
import pandas as pd
import yfinance as yf
from datetime import datetime, timedelta
import requests

url = "https://en.wikipedia.org/wiki/List_of_BSE_SENSEX_companies"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
tables = pd.read_html(response.text)

sensex_constituents = tables[0]

sensex_constituents['YahooSymbol'] = sensex_constituents['Ticker'].astype(str) + ".BO"

sensex_constituents.to_csv('sensex_constituents.csv', index=False)
print("Saved → sensex_constituents.csv")


start_date = (datetime.now() - timedelta(days=10*365)).strftime('%Y-%m-%d')

print("\nFetching 10 years of daily data...\n")
dfs = []

for symbol in sensex_constituents['YahooSymbol']:
    try:
        print(f"Downloading → {symbol}")
        df = yf.download(symbol, start=start_date)
        df = df[['Close']]
        df.columns = [symbol]
        dfs.append(df)
    except Exception as e:
        print(f"Failed → {symbol}, Reason: {e}")

sensex_data = pd.concat(dfs, axis=1)
sensex_data.to_csv("sensex_data.csv")
print("\nSaved → sensex_data.csv")


df = pd.read_csv("sensex_data.csv", index_col=0)
df_filled = df.ffill()

print("\nPreview:")
print(df_filled.head())
```

```python
daily_returns = df_filled.pct_change()
daily_returns.head()
```

```python
biggest_losers = daily_returns.apply(lambda day: day.nsmallest(10).index, axis=1)
biggest_losers.head()
```

```python
capital = 10000

capital_over_time = pd.DataFrame(index=daily_returns.index)


for i in range(len(daily_returns.index) - 1):
    losers = biggest_losers.iloc[i]
   
    prices_today = df_filled.loc[daily_returns.index[i], losers]
    prices_tomorrow = df_filled.loc[daily_returns.index[i+1], losers]
   
    shares = (capital / 10) / prices_today
    capital = (shares * prices_tomorrow).sum()
   
    capital_over_time.loc[daily_returns.index[i+1], 'Capital'] = capital
capital_over_time.head()
```

```python
import numpy as np


returns = capital_over_time['Capital'].pct_change().dropna()

annualized_return = (1 + returns.mean())**252 - 1

annualized_vol = returns.std() * np.sqrt(252)

sharpe_ratio = annualized_return / annualized_vol

rolling_max = capital_over_time['Capital'].cummax()
daily_drawdown = capital_over_time['Capital'] / rolling_max - 1.0
max_drawdown = daily_drawdown.cummin()

print('Annualized Return:', annualized_return)
print('Annualized Volatility:', annualized_vol)
print('Sharpe Ratio:', sharpe_ratio)
print('Max Drawdown:', max_drawdown.min())
```

```python
import yfinance as yf
import numpy as np

sensex_data = yf.download('HDFCSENSEX.NS', start=daily_returns.index.min(), end=daily_returns.index.max())['Close']

sensex_returns = sensex_data.pct_change().dropna()

sensex_annual_return = (1 + sensex_returns.mean())**252 - 1

sensex_annual_vol = sensex_returns.std() * np.sqrt(252)

sensex_sharpe_ratio = sensex_annual_return / sensex_annual_vol
print("Sensex Sharpe Ratio:", sensex_sharpe_ratio)
sensex_sharpe_ratio = float(sensex_sharpe_ratio)

print("\nComparison:")
if sharpe_ratio > sensex_sharpe_ratio:
    print("Mean reversion strategy outperformed the SENSEX!")
else:
    print("My \"VH\" strategy underperformed compared to the SENSEX benchmark.")
```

```python
performance_metrics = pd.DataFrame({
    'Strategy': ['Mean Reversion', 'SENSEX'],
    'Annualized Return': [annualized_return, sensex_annual_return],
    'Annualized Volatility': [annualized_vol, sensex_annual_vol],
    'Sharpe Ratio': [sharpe_ratio, sensex_sharpe_ratio]
})

print(performance_metrics.to_string(index=False))

if sharpe_ratio > sensex_sharpe_ratio:
    print("The mean reversion strategy had the higher risk-adjusted returns.")
else:
    print("The SENSEX benchmark had the higher risk-adjusted returns.")
```
