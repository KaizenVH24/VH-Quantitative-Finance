```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime as dt
```

```python
enddate = dt.datetime.now()
startdate = enddate - dt.timedelta(days = 365*3)

startdate, enddate
```

```python
stocks = ['INDIGO.NS', 'ITC.NS', 'TCS.NS']
```

```python
df = yf.download(stocks, start=startdate, end=enddate)
```

```python
df.head()
```

```python
close_prices = df['Close']
close_prices.head()
```

```python
log_returns = np.log(close_prices/close_prices.shift(1))
log_returns.head()
```

```python
comulative_log_returns = log_returns.cumsum()
```

```python
import matplotlib as plt
comulative_log_returns.plot(title="Cumulative Returns", figsize=(10,6))
```
