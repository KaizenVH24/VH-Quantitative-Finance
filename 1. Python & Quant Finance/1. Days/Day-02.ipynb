```python
import pandas as pd
import numpy as np
import datetime as dt
import yfinance as yf
import os
import openpyxl
```

```python
stocks = ['TCS.NS', 'ITC.NS', 'INDIGO.NS', 'NTPC.NS', 'MARUTI.NS']
```

```python
enddate = dt.datetime.today()
enddate
```

```python
startdate = enddate - dt.timedelta(days=2*365)
startdate
```

```python
close_prices = pd.DataFrame()
```

```python
for stock in stocks:
    data = yf.download(stock, start=startdate, end=enddate)
    close_prices[stock] = data['Close']
```

```python
output_file = os.path.join(output_path, "D2Stock_prices.xlsx")
close_prices.to_excel(output_file, index=False)
```
